```python
def split_ora(ora):
    return re.findall('\d\d', ora)

def split_gradi(gradi):
    return re.findall('[\-]?\d', gradi)
```

```python
newDF['GradoFix'] = newDF['Grado'].apply(split_gradi)
newDF['OraFix'] = newDF['Ora'].apply(split_ora)
newDF['OraFix2'] = ""

i = 0
while i < len(newDF.index):
    newDF.OraFix2[i] = ''.join(newDF.OraFix[i])
    newDF.GradoFix[i] = ''.join(newDF.GradoFix[i])
    i+=1
```

```python
newDF['OraFix2'] = [x[0:2] + ":" + x[2:4] for x in newDF['OraFix2']] 
```

```python
from datetime import datetime
from datetime import timedelta

newDF['OraFix4'] = ''
i = 0
while i < len(newDF.index):
    newDF.OraFix4[i] = datetime.strptime(newDF['OraFix2'][i],'%H:%M').time()
    i+=1    

newDF['OraFix4'] = newDF.OraFix4 = pd.to_datetime(newDF.OraFix4, format='%H:%M:%S')
newDF['OraFix4'] = newDF['OraFix4'] + timedelta(minutes=10)
newDF['OraFix4'] = [datetime.time(d) for d in newDF['OraFix4']] 
newDF['Ora'] = newDF['OraFix4']
newDF['Grado'] = newDF['GradoFix']  
newDF["Grado"] = newDF["Grado"].astype(int)

del newDF['GradoFix']
del newDF['OraFix'] 
del newDF['OraFix2'] 
del newDF['OraFix4'] 

newDF
```
