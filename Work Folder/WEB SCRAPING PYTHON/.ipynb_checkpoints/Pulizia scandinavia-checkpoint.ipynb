```python
# Load the Pandas libraries with alias 'pd' 
import pandas as pd 

# Read data from file 'filename.csv' 
# (in the same directory that your python process is based)
# Control delimiters, rows, column names with read_csv (see later) 
data1 = pd.read_csv("C:/Users/vizzi/Documents/temperature_toto_2012.csv") 

# Preview the first 5 lines of the loaded data 
data1.head()
```

```python
i = 0
while i < len(data1.index):
    data1.OraFix2[i] = ''.join(data1.OraFix2[i])
    #newDF.GradoFix[i] = ''.join(newDF.GradoFix[i])
    i+=1

patternDel = "\:[2]"
filterx = data1.OraFix2.str.contains(patternDel)

data11 = data1[~filterx]

data11
```

```python
# salvo temperature
data11.to_csv("C:/Users/vizzi/Documents/temperature_toto_2012_pulito.csv", index=False)
```
